```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cufflinks as cf
from IPython.display import display, HTML

cf.set_config_file(sharing = 'public',theme='ggplot',offline=True)
```

```python
# reading the csv file
data = pd.read_csv("population_total.csv")
```

```python
# first look at the dataset
data
```

# 1 make a pivot table

##### .pivot() returns reshaped dataframe orhanized by given / columns values (pivot without aggregation)

```python
# dropping null values
data =data.dropna()
data
```

```python
#making a pivot table
data_pivot = data.pivot(index="year",columns="country",values="population")
data_pivot
```

```python
# selecting some countries
data_pivot = data_pivot[['United States','India','China','Indonesia','Brazil']]
```

```python
# showing pivot table
data_pivot
```

# 2. Lineplot

```python
 data_pivot.plot(kind='line',xlabel='Year',ylabel='Population',title='Population (1955-2020)', figsize=(8,4))

# save plot
plt.savefig('my_lineplot.png')
# show plot
plt.show()
```

# 3. Barplot

#### 3.1 single Barplot

```python
# selecting only one year (2020)data_pivot.
data_pivot.loc[[2020]]  # or data_pivot[data_pivot.index == 2020]
data_pivot[data_pivot.index == 2020]   # OR

dd = data_pivot.query("index == 2020")
```

```python
# showing new dataframes
dd
```

```python
# transpose dataframe from rows to columns or viceversa
dd = dd.T
```

```python
# making barplot
dd.plot(kind='bar',color='orange',ylabel='population',xlabel='countries',title='Population 2020')
plt.show()
```

# 3.2 Barplot grouped by 'n' variables

```python

# selecting a few years
bar = data_pivot.loc[[1980,1990,2000,2010,2020]]
```

```python
# making grouped barplot
bar.plot(kind='bar',xlabel='year',ylabel='population',title='Population (1980 -2020)')
plt.show()
```

# 4 piechart

```python
# changing column name
dd.rename(columns={2020 :'2020'},inplace = True)
```

```python
# making the  plot   y = is column having the data
dd.plot(kind='pie',y='2020',title='Population in 2020(%)')
plt.show()
```

```python
data_pivot.plot(kind='hist')
plt.show()
```
